```python
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np
import scipy.ndimage as nd
```

```python
test = scipy.io.loadmat('chair_000000182_1.mat')
```

```python
def getVoxelFromMat(path, cube_len=64):
    """Mat 데이터로 부터 Voxel 을 가져오는 함수"""
    voxels = sio.loadmat(path)['instance']
    voxels = np.pad(voxels, (1, 1), 'constant', constant_values=(0, 0))
    if cube_len != 32 and cube_len == 64:
        voxels = nd.zoom(voxels, (1, 1, 1), mode='constant', order=0)
    return voxels
```

```python
def plotFromVoxels(voxels):
    z, x, y = voxels.nonzero()
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x, y, z, zdir='z', c='blue')
    plt.show()
```

```python
path = 'chair_000000182_4.mat'
voxel = getVoxelFromMat(path)
plotFromVoxels(voxel)
```
